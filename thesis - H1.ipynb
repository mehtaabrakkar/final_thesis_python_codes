```python
import re

# ----------------------------
# KEYWORD LISTS
# ----------------------------
human_terms = ["participants", "patients", "adults", "children", "cohort", "subjects"]

# ----------------------------
# CVD TERMS
# ----------------------------
cvd_terms = [
    # General
    "cardiovascular disease", "cardiovascular diseases", "cvd",
    "heart disease", "heart diseases", "cardiac disease", "cardiac diseases",

    # Coronary / ischemic
    "coronary artery disease", "cad", "ischemic heart disease", "ihd",
    "myocardial infarction", "heart attack",
    "acute coronary syndrome", "acs", "angina",

    # Heart failure
    "heart failure", "congestive heart failure", "chf",
    "cardiomyopathy",

    # Cerebrovascular
    "stroke", "cerebrovascular disease", "cerebrovascular diseases",
    "cerebrovascular accident", "cva",
    "ischemic stroke", "hemorrhagic stroke",
    "transient ischemic attack", "tia",

    # Arrhythmias
    "arrhythmia", "atrial fibrillation", "afib",
    "ventricular tachycardia", "ventricular fibrillation",

    # Vascular
    "atherosclerosis", "arteriosclerosis",
    "peripheral artery disease", "pad",
    "vascular disease", "thrombosis", "embolism",

    # Risk factors
    "hypertension", "high blood pressure",
    "hyperlipidemia", "dyslipidemia", "cholesterol"
]

# ----------------------------
# RACE / ETHNICITY TERMS
# ----------------------------
black_terms = ["black", "african american"]

south_asian_terms = [
    "south asian", "indian", "pakistani", "bangladeshi", "sri lankan",
    "nepalese", "bhutanese", "desi"
]

# Broad Asian category
asian_agg_terms = ["asian", "asian american", "asian americans"]

# ANY Asian subgroup = disaggregation (expanded to include less common)
disagg_terms = [
    "indian", "pakistani", "bangladeshi", "sri lankan", "nepalese", "bhutanese",
    "chinese", "japanese", "korean", "vietnamese", "filipino", "thai",
    "malaysian", "singaporean", "indonesian", "cambodian", "laotian", "burmese"
]

# ----------------------------
# FUNCTIONS
# ----------------------------

# Case-insensitive + word boundaries for short abbreviations
def check_terms(text, terms):
    if not isinstance(text, str):
        return False
    text = text.lower()
    for term in terms:
        if term in ["cad", "acs", "chf", "cva", "tia", "afib", "pad", "ihd", "cvd"]:
            if re.search(rf"\b{term}\b", text):
                return True
        else:
            if term in text:
                return True
    return False

# Fixed US detection
def check_us(text):
    if not isinstance(text, str):
        return False
    return (
        "united states" in text.lower() or
        "usa" in text.lower() or
        "u.s." in text.lower() or
        re.search(r"\bUS\b", text, re.IGNORECASE) is not None or
        "american" in text.lower()
    )

# ----------------------------
# APPLY CODING
# ----------------------------
df["US_based"] = df["Abstract"].apply(check_us)

df["Human_subjects"] = df["Abstract"].apply(lambda x: check_terms(x, human_terms))
df["CVD_flag"] = df["Abstract"].apply(lambda x: check_terms(x, cvd_terms))
df["Black_mentions"] = df["Abstract"].apply(lambda x: check_terms(x, black_terms))
df["South_Asian_mentions"] = df["Abstract"].apply(lambda x: check_terms(x, south_asian_terms))
df["Asian_aggregation"] = df["Abstract"].apply(lambda x: check_terms(x, asian_agg_terms))
df["Disaggregation"] = df["Abstract"].apply(lambda x: check_terms(x, disagg_terms))

# ----------------------------
# SAVE
# ----------------------------
df.to_csv("coded_abstracts.csv", index=False)
print("✅ Done! Saved as coded_abstracts.csv")
```

```python
import pandas as pd
import re

def parse_pubmed_file(filename):
    records = []
    current_record = {"Abstract": "", "Affiliations": ""}

    with open(filename, "r") as file:
        for line in file:
            line = line.strip()
            
            if line.startswith("PMID-"):
                if current_record["Abstract"] or current_record["Affiliations"]:
                    records.append(current_record)
                current_record = {"Abstract": "", "Affiliations": ""}
            
            elif line.startswith("AB  -"):
                current_record["Abstract"] += line.replace("AB  -", "").strip() + " "
            
            elif line.startswith("AD  -"):
                current_record["Affiliations"] += line.replace("AD  -", "").strip() + " "
        
        records.append(current_record)

    return pd.DataFrame(records)
```

```python
import re
import pandas as pd

def parse_pubmed_file_with_pmid(filename):
    records = []
    current_record = {"PMID": "", "Abstract": "", "Affiliations": ""}

    with open(filename, "r") as file:
        for line in file:
            line = line.strip()
            
            # New record starts with PMID
            if line.startswith("PMID-"):
                if current_record["PMID"]:
                    records.append(current_record)
                current_record = {"PMID": "", "Abstract": "", "Affiliations": ""}
                current_record["PMID"] = line.replace("PMID-", "").strip()
            
            elif line.startswith("AB  -"):
                current_record["Abstract"] += line.replace("AB  -", "").strip() + " "
            
            elif line.startswith("AD  -"):
                current_record["Affiliations"] += line.replace("AD  -", "").strip() + " "

    # Add last record
    records.append(current_record)
    return pd.DataFrame(records)
```

```python
# Group by year
yearly = df.groupby("Publication Year").agg(
    total_included=("Final_inclusion", "sum"),
    black_mentions=("Black_mentions", "sum"),
    sa_mentions=("South_Asian_mentions", "sum"),
    asian_agg=("Asian_aggregation", "sum"),
    disagg=("Disaggregation", "sum")
).reset_index()

# Compute proportions
yearly["black_pct"] = yearly["black_mentions"] / yearly["total_included"]
yearly["sa_pct"] = yearly["sa_mentions"] / yearly["total_included"]
yearly["asian_agg_pct"] = yearly["asian_agg"] / yearly["total_included"]
yearly["disagg_pct"] = yearly["disagg"] / yearly["asian_agg"]  # among Asian-focused

print(yearly.head())
```
